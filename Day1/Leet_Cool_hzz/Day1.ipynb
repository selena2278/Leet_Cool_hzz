2. Arrays are stored in contiguous memory locations, which means that adding or 
removing elements typically requires shifting the positions of other elements 
within the memory.

Question 1: Leetcode 704. Binary Search

```python
from typing import List

#[left, right]
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left, right = 0, len(nums)-1
        
        while left <= right:
            middle = left + (right-left)//2

            if nums[middle] > target:
                right = middle-1
            elif nums[middle] < target:
                left = middle+1
            else:
                return middle
        
        return -1

```
